```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
a = 0.65343
b = 0.734535
num = 100000000
xs = np.empty(num+1)
ys = np.empty(num+1)
xs[0] = 1.0
ys[0] = 1.0
```

```python
def fractal(x,y):
  xdot = math.sin(x*y/b) + math.cos(a*x-y)
  ydot = x + math.sin(y)/b
  return xdot, ydot
```

```python
for i in range(num):
  xdot, ydot = fractal(xs[i], ys[i])
  xs[i+1]=xs[i]+xdot
  ys[i+1]=ys[i]+ydot
```

```python
ax = plt.figure()
```

```python
plt.plot(xs,ys)
```
