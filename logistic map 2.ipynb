```python
import numpy as np
import matplotlib.pyplot as plt
R = np.linspace(1.5,4,10000)
print(R)
```

```python
X = []
Y = []
```

```python
for r in R:
  X.append(r)

  x = np.random.random()
  for n in range(100):
    x = r*x*(1-x)

  #for n in range(100):
    #x=r*x*(1-x)

  Y.append(x)
print(X)
print(Y)
```

```python
plt.plot(X,Y, ls='', marker = ',')
plt.show()
```
