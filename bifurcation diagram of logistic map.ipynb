```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.linspace(2,4,100000)
X = []
Y = []
for r in R:
  X.append(r)

  x = np.random.random()
  for n in range(100):
    x = r*x*(1-x)

  #for n in range(100):
    #x=r*x*(1-x)

  Y.append(x)
plt1 = plt.figure(figsize = (14,8))
plt.xlabel('Control Parameter r')
plt.ylabel('Values of x')
plt.plot(X,Y, ls='', marker = ',')
plt.show()
```
