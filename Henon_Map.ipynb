```python
import matplotlib.pyplot as plt
```

```python
def Henonmap(x,y,a,b,size):
  X = []
  Y = []
  x1=y1=0
  for i in range(size):
    x1=1-a*x*y*x+y
    y1=b*x
    print(x1,"",y1)
    x=x1
    y=y1
    X.append(x)
    Y.append(y)
  return X,Y
```

```python
xc, yc = Henonmap(0.001,0.2,1.4,0.3,10)
plt.plot(xc,yc)
plt.show()
```
