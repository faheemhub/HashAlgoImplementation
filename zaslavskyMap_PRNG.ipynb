```python
import math
def ZaslavskyMap_PRNG(x,y,r,v,e,size):
  prn1=prn2=prn3=prn4=0
  x1=y1=0

  u = (1-pow(e,-r))/r

  for i in range(size):

    x1 = ((x+v*(1+u*v)+e*v*u*math.cos(2*math.pi*x))%1)
    y1 = pow(e,-r)*(y+e*math.cos(2*math.pi*x))

    print("\nchaotic values:",'\nx=',x1,"",'\ny=',y1)
    x = x1
    y = y1

    prn1=x1*10000-math.floor(x1*10000)
    print('\nprn1=',prn1)

    prn2=(math.floor(x1*(math.pow(2,35)-1)))%256
    print('\nprn2=',prn2)

    prn3=int(math.floor(y1*(math.pow(2,20)-1)))%1024
    print('\nprn3=',prn3)

    prn4=("{0:b}".format(prn3))
    print('\nprn4-binary form:',prn4)
```

```python
x=0.1
y=0.2
v=4
e=2.3
r=3
size=1
ZaslavskyMap_PRNG(x,y,r,v,e,size)
```
